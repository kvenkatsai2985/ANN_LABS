```python
def binaryStep(x):
    return np.heaviside(x,1)
```

```python
import matplotlib.pyplot as plt
x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Activation Function :binaryStep')
plt.show()
```

```python
def linear(x):
    return x
```

```python
x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Activation Function :Linear')
plt.show()
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10,10);
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
def tanh(x):
  return np.tanh(x)
```

```python
x = np.linspace(-10,10);
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation function :tanh')
plt.show()
```

```python
def RELU(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :RELU')
plt.show()
```
